# Game: Hangman (Guess the Word)

This notebook contains a simple Hangman game created with Python basics. It uses lists, string manipulation, while loops, and user input.

The game picks a secret word and you guess letters to reveal it. Get it right before you run out of guesses!

```python
# Game: Hangman (Guess the Word)
# Created with Copilot assistance
# Learned: Lists, strings, while loops, string methods, input validation

def play_hangman():
    """Main Hangman game function"""
    # Word list for the game
    words = ["python", "hangman", "computer", "programming", "notebook", "copilot"]
    
    # Pick a random word
    import random
    secret_word = random.choice(words)
    
    # Game variables
    guessed_letters = []
    wrong_guesses = 0
    max_wrong = 6
    
    print("Welcome to Hangman!")
    print(f"The word has {len(secret_word)} letters.")
    print()
    
    # Main game loop
    while wrong_guesses < max_wrong:
        # Display current state of word
        display = ""
        word_complete = True
        
        for letter in secret_word:
            if letter in guessed_letters:
                display += letter
            else:
                display += "_"
                word_complete = False
        
        print(f"Word: {display}")
        print(f"Guessed letters: {', '.join(guessed_letters)}")
        print(f"Wrong guesses: {wrong_guesses}/{max_wrong}")
        print()
        
        # Check if player won
        if word_complete:
            print(f"You won! The word was: {secret_word}")
            break
        
        # Get player's guess
        guess = input("Guess a letter: ").lower()
        
        # Validate input
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue
        
        if guess in guessed_letters:
            print("You already guessed that letter!")
            continue
        
        # Add guess to list
        guessed_letters.append(guess)
        
        # Check if guess is correct
        if guess in secret_word:
            print("Good guess!")
        else:
            print("Wrong guess!")
            wrong_guesses += 1
        
        print()
    
    # Check if player lost
    if wrong_guesses >= max_wrong:
        print(f"Game Over! The word was: {secret_word}")

play_hangman()
```
